```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create data for three models
model1_data = {'Model': ['Cosine similarity'], 'Metric': ['Accuracy'], 'Value': [0.73]}
model2_data = {'Model': ['Kmeans'], 'Metric': ['Accuracy'], 'Value': [0.69]}
model3_data = {'Model': ['SVM'], 'Metric': ['Accuracy'], 'Value': [0.43]}

# Create Pandas DataFrames for each model's data
df_model1 = pd.DataFrame(data=model1_data)
df_model2 = pd.DataFrame(data=model2_data)
df_model3 = pd.DataFrame(data=model3_data)
# Set the figure size and style for the Seaborn plot
#sns.set(rc={'figure.figsize':(10,6)})
sns.set_style('whitegrid')

# Create a Seaborn barplot to show the accuracy and precision of each model
sns.barplot(x='Model', y='Value', hue='Metric', data=pd.concat([df_model1, df_model2, df_model3]), palette='Set2')

# Add labels and title to the plot
plt.xlabel('Model')
plt.ylabel('Value')
plt.title('Comparison of Accuracy  for Music Recommendation Models')

# Show the plot
plt.show()






```
